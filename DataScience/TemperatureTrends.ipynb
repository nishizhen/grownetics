# Data Visualization

Load some data from CSV, then plot it on a graph.

```python
import matplotlib.pyplot as plt
import csv
from datetime import datetime
from dateutil.parser import parse

hum_dates = []
temp_dates = []
hums = []
temps = []

# 3 months of data, a fair bit
#with open('CSVs/2019-04-03-12-19 Chronograf Data.csv','r') as csvfile:
    
# Less data, faster
with open('CSVs/2019-04-03-12-17 Chronograf Data.csv','r') as csvfile:
    plots = csv.reader(csvfile, delimiter=',')
    for row in plots:
       # print(row)
        if (len(row[1]) and row[2].isnumeric()):
            if (row[3] == '2'):
                hum_dates.append(parse(row[0]))
                hums.append(float(row[1]))
            else:
                temp_dates.append(parse(row[0]))
                temps.append((float(row[1]) * 9/5) + 32)


fig = plt.figure(figsize=(18,10))
#print(dates)
plt.plot(temp_dates, temps, label='Temperature')
plt.plot(hum_dates, hums, label='Humidity')
plt.xlabel('Date')
plt.ylabel('Value')
plt.title('Temperature Trends')
plt.legend()
plt.show()
```
